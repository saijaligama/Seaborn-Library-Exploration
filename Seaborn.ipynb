```python
import seaborn as sns
```

```python
crash_df = sns.load_dataset('car_crashes')
```

```python
crash_df.head()
```

# distribution plots, plot single variable

```python
# sns.distplot(crash_df['not_distracted']) # kde = false if we dont want the line
sns.distplot(crash_df['not_distracted'],bins=25)
```

# joint plot for 2 distributions

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data=crash_df,kind='reg')
```

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data=crash_df,kind='kde')
```

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data=crash_df,kind='hex')
```

# Styling

```python
import matplotlib.pyplot as plt
sns.set_style('white') # other options 'whitegrid' , 'dark','darkgrid','ticks'

plt.figure(figsize=(8,4)) #plot size

sns.set_context('paper',font_scale=1.4)

sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')

sns.despine(left=True) # removes the left axis other options bottom,top,rightm
```

```python
sns.set_style('white') # other options 'whitegrid' , 'dark','darkgrid','ticks'

plt.figure(figsize=(8,4)) #plot size

sns.set_context('talk',font_scale=1.4)

sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

```python
sns.set_style('white') # other options 'whitegrid' , 'dark','darkgrid','ticks'

plt.figure(figsize=(8,4)) #plot size

sns.set_context('poster',font_scale=1.4) # essentially scaling options

sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

# pairgrids

```python
iris = sns.load_dataset('iris')
iris_grid = sns.PairGrid(iris,hue='species')
iris_grid.map(plt.scatter)
```

# use multiple type of plots in a single pair plot

```python
iris_grid = sns.PairGrid(iris,hue='species')
iris_grid.map_diag(plt.hist)
iris_grid.map_upper(plt.scatter)
iris_grid.map_lower(sns.kdeplot)
```
